# üìà Evaluating the Option Price

To calculate the theoretical fair value of the option under the Black-Scholes model, simply call the `evaluate()` method. This method requires no input arguments, as it relies on the parameters provided during initialization (e.g., spot price, strike, time to maturity, volatility, etc.).

Once called, it returns the Black-Scholes price of the option ‚Äî either a **Call** or **Put**, depending on the specified `option_type`.

````python
bs.evaluate()
````

## ‚öôÔ∏è `d1` and `d2` Parameters

The `d1` and `d2` properties return the intermediate terms used in the Black-Scholes formula. These values are essential components in the computation of the option's price and Greeks.

- `d1` reflects the standardized distance between the current price and the strike price, adjusted for volatility and time.
- `d2` is derived directly from `d1` and represents the standardized distance accounting for the uncertainty of the asset's return.

These properties are computed automatically based on the model‚Äôs internal parameters (spot price, strike, volatility, interest rate, time to maturity, and dividend yield) and do not require any external input.

````python
d1, d2 = bs.d1, bs.d2
print(f'd1 value is: {d1}')
print(f'd2 value is: {d2}')
````

## üßÆ Greeks

The `BlackScholes` class provides built-in support for calculating five key sensitivity measures - commonly referred to as the *Greeks* - which describe how the option price responds to changes in market parameters:

- **Delta (Œî):** Measures the sensitivity of the option price to changes in the underlying asset's price.
- **Gamma (Œì):** Measures the rate of change of Delta with respect to changes in the underlying price (i.e., the curvature).
- **Vega (ŒΩ):** Measures sensitivity to changes in the implied volatility of the underlying asset.
- **Theta (Œò):** Measures the rate of decline in the option‚Äôs value due to the passage of time (time decay).
- **Rho (œÅ):** Measures sensitivity to changes in the risk-free interest rate.

Each of these Greeks is available as a dedicated method and is calculated analytically using the closed-form formulas derived from the Black-Scholes model.

These metrics are essential for risk management, hedging strategies, and understanding the behavior of options under different market scenarios.


````python
delta = bs.compute_delta()
gamma = bs.compute_gamma()
rho = bs.compute_rho()
theta = bs.compute_theta()
vega = bs.compute_vega()
````

````python
print(f'The delta (Œî) value is: {delta}')
print(f'The gamma (…£) value is: {gamma}')
print(f'The rho (œÅ) value is: {rho}')
print(f'The theta (Œ∏) value is: {theta}')
print(f'The vega (ŒΩ) value is: {vega}')
````

## üìä Summary

To retrieve all the key outputs of the model at once ‚Äî including the option price and its associated Greeks ‚Äî use the `summary()` method.

This method returns a structured dictionary containing:
- The theoretical **option price**
- All five major **Greeks**: Delta, Gamma, Vega, Theta, and Rho

By default, the method returns a standard Python dictionary. If you prefer a tabular format, set the `as_frame=True` parameter to receive the results as a **Pandas DataFrame**, making it easier to visualize or export.

> ‚ÑπÔ∏è There is no need to call `evaluate()` beforehand ‚Äî `summary()` performs all required calculations internally.

````python
summary = bs.summary(as_frame=True)
summary
````

### üìà Sensitivity Analysis

The `BlackScholes` class includes three specialized methods designed to evaluate how the **option price** and its **Greeks** respond when a single parameter is varied, while all others remain constant:

- `greeks_over_time(T_range)`  
  Varies **time to maturity**. Useful for understanding how time decay (Theta) and other sensitivities evolve as expiration approaches.

- `greeks_over_price(St_range)`  
  Varies the **underlying asset price**. Particularly relevant for analyzing Delta and Gamma around the strike price.

- `greeks_over_volatility(sigma_range)`  
  Varies **implied volatility**. Highlights the behavior of Vega and the non-linear effects of volatility on the option price.

#### Input:
Each method accepts a single argument ‚Äî a **range of strictly positive values** (list, NumPy array, or similar) corresponding to the parameter being varied.

#### Output:
All three methods return a **Pandas DataFrame** where:
- Each **row** corresponds to one value from the input range  
- Each **column** contains:
  - The **option price**
  - The five primary **Greeks**: Delta, Gamma, Vega, Theta, and Rho

This format is ideal for further analysis, visualization, or exporting to external tools.

> ‚ö†Ô∏è **Important:** All input values must be strictly positive. Negative or zero values are not valid in the Black-Scholes model and will raise errors or produce invalid results.

````python
def plot_summary(df, title=None):
    fig, ax = plt.subplots(figsize=(16,8))
    ax.plot(df)
    ax.legend(df.columns)
    ax.grid(ls='--')
    if title:
        fig.suptitle(title)
    sns.despine()
    plt.show()
````

### 1. Greeks Over Price

````python
srange = np.linspace(190, 220, 100)
gop = bs.greeks_over_price(srange)
gop.head()
````

````python
plot_summary(gop.drop(columns='option price'),
            title='Greeks Change Over the Stock Price')
````

### 2. Greeks Over Time

````python
trange = np.linspace(0.001, 1, 100)
got = bs.greeks_over_time(trange)
got.head()
````

````python
plot_summary(got.drop(columns='option price'),
            title='Greeks Change Over Time')
````

### 3. Greeks Over Volatility

````python
vrange = np.linspace(0.001, 0.5, 100)
gov = bs.greeks_over_volatility(vrange)
gov.head()
````

````python
plot_summary(gov.drop(columns='option price'),
            title='Greeks Chagne Over Volatility')
````

## üåê Volatility Surface

The `volatility_surface()` method allows you to compute the **implied volatility surface** over a grid of strike prices and times to maturity, given a known market price for the option.

This surface provides critical insights into how the **implied volatility** varies across different strike levels and expirations ‚Äî a foundational tool in derivatives analysis, risk management, and volatility trading.

#### Parameters:
- `market_price` (`float`): The observed market price of the option (e.g., from an options chain).
- `K_range` (`array-like`): A range of **strike prices** to evaluate. Must consist of positive values.
- `T_range` (`array-like`): A range of **time-to-maturity values** (in years). Must consist of positive values.
- `n_jobs` (`int`, optional): Number of CPU cores to use for **parallel computation**. Default is 1 (sequential); set to a higher value (e.g., 4, 8, 12) for faster performance on large grids.

#### Returns:
- A **2D NumPy array** of implied volatility estimates, with shape `(len(K_range), len(T_range))`, suitable for visualization using contour or surface plots.


````python
krange = np.linspace(200, 220, 100)
trange = np.linspace(0.01, 0.2, 100)
mp = call27['lastPrice']
vol_surface = bs.volatility_surface(market_price=mp,
                                    K_range=krange,
                                    T_range=trange,
                                   n_jobs=12)
````

````python
vol_surface
````

### üõ∞Ô∏è Volatility Surface Plotting

After computing the implied volatility surface using the `volatility_surface()` method, you can visualize it using the `plot_volatility_surface()` method.

This method generates a **3D surface plot** of implied volatility across a grid of strike prices and times to maturity ‚Äî a powerful visual tool for identifying patterns such as volatility smiles, skews, or term structure effects.

#### Parameters:
- `elev` (`float`, optional): Controls the **elevation angle** of the 3D plot (in degrees). Affects the vertical viewing angle.  
- `azim` (`float`, optional): Controls the **azimuthal angle** of the plot (in degrees). Affects the horizontal rotation around the plot.  
- `cmap` (`str`, optional): The colormap used for the surface. Default is `'OrRd'`, but any valid Matplotlib colormap name can be used (e.g., `'viridis'`, `'coolwarm'`, `'plasma'`).

These parameters allow full control over the appearance and orientation of the surface for better interpretability or presentation purposes.

#### Output:
The method renders a **Matplotlib 3D surface plot** and does not return a value.

> ‚ÑπÔ∏è Ensure that the volatility surface has been computed beforehand. The plotting method assumes that valid `K_range`, `T_range`, and surface data are already available.

````python
bs.plot_volatility_surface()
````

### üìå Implied Volatility

The `implied_volatility()` method allows you to compute the **implied volatility** of an option based on its current **market price**.

Implied volatility is the level of volatility that, when plugged into the Black-Scholes formula, yields the given market price. It is a reverse-engineering process, often used by traders to infer market expectations of future volatility.

#### Usage:
To use this method, you must supply the **current market price** of the option as an argument:
```python
iv = bs.implied_volatility(market_price=5.40)
```

#### Parameters:
- `market_price` (`float`): The observed price of the option in the market. Must be a positive value.

#### Returns:
- A single **float** representing the implied volatility (annualized), typically expressed in decimal form (e.g., `0.24` for 24%).

> ‚ö†Ô∏è This method uses numerical root-finding (e.g., Brent‚Äôs method) under the hood. It may fail or return `np.nan` if the market price is not within a plausible range given the model‚Äôs assumptions.

````python
iv = bs.implied_volatility(mp)
print(f'The implied volatility is: {round(iv, 3)}')
````
